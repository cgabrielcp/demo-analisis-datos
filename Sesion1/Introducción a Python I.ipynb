#### Tuple Data Type
Una tupla en Python es similar a una lista. La diferencia entre los dos es que no podemos cambiar los elementos de una tupla una vez asignada, mientras que pen una lista si lo podemos hacer.

```python
# Diferentes tipos de tuplas

# Tuple Empty
my_tuple = ()
print(my_tuple)

# Tuple int
my_tuple = (3, 2, 5)
print(my_tuple)

# Mixed Tuple
my_tuple = (1, "Hello", 5.6)
print(my_tuple)

# Nested Tuple
my_tuple = ("apple", [2, 5, 7], (5, 8, 1))
print(my_tuple)
```

```python

my_tuple = ("hello")
print(type(my_tuple))  # <class 'str'>

# Tupla con un elemento
my_tuple = ("hello",)
print(type(my_tuple))  # <class 'tuple'>

# el parentesis es opcional
my_tuple = "hello","world","people"
print(type(my_tuple))  # <class 'tuple'>
```

#### Dictionary Data Type
El diccionario de Python es una colección desordenada de elementos. Cada elemento de un diccionario tiene un par clave/valor.

Los diccionarios están optimizados para recuperar valores cuando se conoce la clave.

Crear un diccionario es tan simple como colocar elementos entre llaves { } separados por comas.

Un elemento tiene una clave y un valor correspondiente que se expresa como un par (clave: valor).

Si bien los valores pueden ser de cualquier tipo de datos y pueden repetirse, las claves deben ser de tipo inmutable (cadena, número o tupla con elementos inmutables) y deben ser únicas.


```python
# Diccionario Vacío
my_dict = {}

# Diccionario con una llave entera
my_dict = {
            1: 'apple', 
           2: 'pera'}

# use dict()
my_dict = dict({1:'apple', 2:'pera'})
my_dict
```

```python
# Diccionario con distintos tipos de llaves
my_dict_2 = {
                'name': 'John', 
                1: {"edad":12}
            }
my_dict_2
```

```python
# While

n = 10 ## bandera

# inicializamos el contador y la variable suma
suma = 0
# suma = 0, 1, 3..
# i = 1, 2, 3...10, 11

while i <= n:
    suma = suma + i
    i+=1    # actualizar el contador


# print sum
print("The sum is", suma)
```

Ejemplo de la base de datos transaccional

```python
class Colaborador():
  def __init__(self, nombres,dni, direccion, telefono):
    self.nombres=nombres
    self.direccion=direccion
    self.telefono=telefono
    self.dni=dni
  def saludaColaborador(self):
      return f"Hola, soy colaborador en datapath y me llamo {self.nombres}"
```

```python
class Coordinador(Colaborador):
  def __init__(self, nombres,dni, direccion, telefono, fecha_integro):
    super().__init__(nombres,dni, direccion, telefono)
    self.fecha_integro=fecha_integro
  def saludaCoordinador(self):
    return f"Hola, soy Coordinador y me llamo: {self.nombres}"
```

```python
class Docente(Colaborador):
  def __init__(self, nombres,dni, direccion, telefono, programaAsignado):
    super().__init__(nombres,dni, direccion, telefono)
    self.programaAsignado=programaAsignado
  def saludaDocente(self):
    return f"Hola, soy Docente y me llamo: {self.nombres}"
```

### 1. Creación de un DataFrame a partir de Listas

Las dos estructuras de datos principales dentro del paquete Pandas son:

- Series: array unidimensional etiquetado capaz de almacenar cualquier tipo de dato.
- DataFrame: estructura bidimensional con columnas que pueden ser también de cualquier tipo. Estas columnas son a su vez Series.

```python
import pandas as pd
```

```python
# Creación de un DataFrame a partir de Listas

#Frameworks para desarollo web, por lenguaje de
# programación.

frameworks= [
    ["Python","Django"],
    ["PHP","Laravel"],
    ["Java","Spring"],
    ["JavaScript","Angular"]
]


# Creación del DataFrame
df_frameworks=pd.DataFrame(
            frameworks,
            columns=["Lenguaje de programación",
                     "Framework"
                     ]
    )

#visualizamos en DataFrame
df_frameworks.head()

```
